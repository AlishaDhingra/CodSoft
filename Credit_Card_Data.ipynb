```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/creditcard (2).csv')
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
# Data Visualization: Time vs. Amount
plt.figure(figsize=(12, 6))
plt.scatter(df['Time'], df['Amount'], alpha=0.5)
plt.title('Time vs. Amount')
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

```python
# Data Visualization: Correlation Matrix
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Data Visualization: Distribution of features
plt.figure(figsize=(20, 20))
for i, col in enumerate(df.columns[1:29], 1):
    plt.subplot(7, 4, i)
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
